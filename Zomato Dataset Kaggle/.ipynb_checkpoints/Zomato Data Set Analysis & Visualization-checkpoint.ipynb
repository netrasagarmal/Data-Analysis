#### Drop very less important columns

```python
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

```python
df.columns
```

#### Drop duplicate rows

```python
df.drop_duplicates(inplace = True)
df.shape
```

#### First clean rating column

```python
df['rate'].unique()
```

we can see some string values like NEW, - and also ratings have /5 joined so these things need to be romeved

```python
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

we can see there are many null values, we can replace them with mean of remaining values, by calculating mean

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.isnull().sum()
```

above we can see there are few hundresd of null vales. so we can drop thode rows having null values.

```python
df.dropna(inplace = True)
df.isnull().sum()
```

now we have got all columns with no null values.


#### Now we will do the further data pre processing

```python
df.head()
```

some columns have very lengthy names, we can make them short by renaming them

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type', 'listed_in(city)':'City'}, inplace = True)
df.head()
```

There are two columns having locations, so we can drop one of them

```python
df.location.unique()
```

```python
df.City.unique()
```

we will drop City column

```python
df = df.drop(['City'], axis = 1)
df.head(2)
```

#### Analyze and clean rest_type column


first ceck unique values count in this column

```python
rest_types = df['rest_type'].value_counts(ascending  = False)
rest_types
```

take out values having count less than 1000, cluster them .ie. group them and make them as other type

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
        
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

#### Now we will work on Cost2Plate column 

so first check unique values in this column

```python
df['Cost2plates'].unique()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```

we have cleaned Cost2Plate column

#### Now lets move to location column

```python
df['location'].value_counts(ascending  = False)
```

we can see some locations have ver less count, so again we an make clusters of locations having count less than 250 and make them as others

```python
location = df['location'].value_counts(ascending  = False)

location_lessthan250 = location[location<250]



def handle_location(value):
    if(value in location_lessthan250):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

#### cuisines and type column are left

#### first consider cuisines column

```python
df['cuisines'].value_counts(ascending  = False)
```

again we can see so cuisines have very less so will repeat the same process again making as others

```python
cuisines = df['cuisines'].value_counts(ascending  = False)


cuisines_lessthan100 = cuisines[cuisines<100]



def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value
        
df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

### Book table with location

```python
df2 = df.groupby(['location','book_table'])['name'].count()
df2.to_csv('location_booktable.csv')
df2 = pd.read_csv('location_booktable.csv')
df2 = pd.pivot_table(df2, values=None, index=['location'], columns=['book_table'], fill_value=0, aggfunc=np.sum)
df2
```

```python
df2.plot(kind = 'bar', figsize = (15,8))
```

### Online Order Facility, Location Wise

```python
df1.plot(kind = 'bar', figsize = (15,8))
```

### No of votes vs location

```python
df4 = df[['location', 'votes']]
df4.drop_duplicates()
df5 = df4.groupby(['location'])['votes'].sum()
df5 = df5.to_frame()
df5 = df5.sort_values('votes', ascending=False)
df5.head()
```

```python
plt.figure(figsize = (15,8))
sns.barplot(df5.index , df5['votes'])
plt.xticks(rotation = 90)
```
